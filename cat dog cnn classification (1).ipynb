```python
import numpy as np
import cv2
import random
import matplotlib.pyplot as plt
import pickle
import os
```

```python
DIRECTORY=r'C:\Users\SHUBHAM\Desktop\CNNCD\dogscats\dogscats\train'
CATEGORIES=['cats','dogs']
```

```python
IMG_SIZE=70
data=[]
for category in CATEGORIES:
    folder=os.path.join(DIRECTORY,category)
    label= CATEGORIES.index(category)
    for img in os.listdir(folder):
        img_path=os.path.join(folder,img)
        img_arr=cv2.imread(img_path)
        img_arr=cv2.resize(img_arr,(IMG_SIZE,IMG_SIZE))
        data.append([img_arr,label])
```

```python
random.shuffle(data)
```

```python
x=[]
y=[]


for features,labels in data:
    x.append(features)
    y.append(labels)
```

```python
x=np.array(x)
y=np.array(y)
```

```python
pickle.dump(x,open('x.pkl','wb'))
pickle.dump(y,open('y.pkl','wb'))
```

```python
x=pickle.load(open('x.pkl','rb'))
y=pickle.load(open('y.pkl','rb'))
```

```python
x= x/255
```

```python
x.reshape(-1,70,70,1)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPooling2D,Flatten,Dense
```

```python
model= Sequential()


model.add(Conv2D(64,(3,3),activation='relu'))
model.add(MaxPooling2D((2,2)))
model.add(Flatten())


model.add(Dense(128,input_shape=(70,70,3),activation='relu'))
model.add(Dense(2,activation='softmax'))
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x,y,epochs=3,validation_split=0.1,batch_size=32)
```

```python
model.save('catdog.model')
```
